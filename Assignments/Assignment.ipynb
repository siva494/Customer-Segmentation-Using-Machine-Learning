##8.Transformation(Encoding,Scaling)

```python
from sklearn.preprocessing import LabelEncoder,StandarScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
le=LabelEncoder()
```

```python
df['CarName']=le.fit_transform(df['CarName'])
df['fueltype']=le.fit_transform(df['fueltype'])
df['aspiration']=le.fit_transform(df['aspiration'])
df['doornumber']=le.fit_transform(df['doornumber'])
df['carbody']=le.fit_transform(df['carbody'])
df['drivewheel']=le.fit_transform(df['drivewheel'])
df['enginelocation']=le.fit_transform(df['enginelocation'])
df['enginetype']=le.fit_transform(df['enginetype'])
df['cylindernumber']=le.fit_transform(df['cylindernumber'])
```

```python
df.drop(columns=['car_ID'])
df
```

## X & y split

```python
X=df.drop(columns=['price'])
y=df['price']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
X=pd.get_dummies(X)
scaled_x=pd.DataFrame(scale.fit_transform(X),columns=X.columns)
scaled_x.head()
```

###9.Performing Train Test split

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,random_state=42)
X_train.head()
```

##10.Train the data with suitable ML model

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

##11.Test the model

```python
y_pred=model.predict(X_test)
y_pred
```

```python
X_test
```

```python
y_test
```

##12.Evaluate the model

```python
#MSE
mse=mean_squared_error(y_test,y_pred)
#R-square
r2=r2_score(y_test,y_pred)
```

```python
mse
```

```python
r2
```

```python
#RMSE
rmse=np.sqrt(mse)
```
